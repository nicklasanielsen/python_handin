```python
import matplotlib.pyplot as plt

def get_data_frame(url):
    return pd.read_csv(url, delimiter = ";")

# Task 5 A
# What is the change in pct of divorced danes from 2008 to 2020?
def task_5_a() :
    print("Task 5 A")

    url = "https://api.statbank.dk/v1/data/FOLK1A/CSV?delimiter=Semicolon&CIVILSTAND=F&Tid=2008K1%2C2020K4"
    df = get_data_frame(url)

    d_2008 = df.iloc[0,2]
    d_2020 = df.iloc[1,2]
    diff_in_pct = round(d_2020 / d_2008 * 100, 2)

    print("2008: {d_2008}\n2020: {d_2020}\nDiff: {diff}%".format(d_2008 = d_2008, d_2020 = d_2020, diff = diff_in_pct))

# Task 5 B
# Which of the 5 biggest cities has the highest percentage of 'Never Married' in 2020?
def task_5_b():
    print("Task 5 B")

    url = "https://api.statbank.dk/v1/data/FOLK1A/CSV?delimiter=Semicolon&Tid=2020K4&OMR%C3%85DE=101%2C461%2C630%2C751%2C851&CIVILSTAND=TOT%2CU"
    df = get_data_frame(url)
    df = df.sort_values(by = "OMRÅDE")

    areas = []

    for index in range(0, len(df.index), 2):
        area_name = df.iloc[index, 1]
        area_population = df.iloc[index, 3]
        area_unmarried = df.iloc[index + 1, 3]
        area_unmarried_pct = round(area_unmarried / area_population * 100, 2)

        area = {"area": area_name, "population": area_population, "unmarried": area_unmarried, "pct": area_unmarried_pct}
        areas.append(area)

    areas = sorted(areas, key = lambda x: x["pct"], reverse = True)
    print(areas[0])

# Task 5 C
# Show a bar chart of changes in marrital status in Copenhagen from 2008 till now
def task_5_c():
    print("Task 5 C")

    url = "https://api.statbank.dk/v1/data/FOLK1A/CSV?timeOrder=Ascending&delimiter=Semicolon&OMR%C3%85DE=101&CIVILSTAND=G&Tid=*"
    df = get_data_frame(url)

    timeline = {}

    for index in range(0, len(df.index)):
        timeline[df.iloc[index, 2]] = df.iloc[index, 3]

    plt.bar(timeline.keys(), timeline.values(), width = 0.5, align = "edge")
    plt.xticks(rotation = 75)
    plt.xlabel("Tid", fontsize = 10)
    plt.ylabel("Antal gifte", fontsize = 10)

# Task 5 D
# Show 2 plots in same figure: 'Married' and 'Never Married' for all ages in DK in 2020
# (Hint: x axis is age from 0-125, y axis is how many people in the 2 categories).
# Add lengend to show names on graphs
def task_5_d():
    print("Task 5 D")

    url = "https://api.statbank.dk/v1/data/FOLK1A/CSV?timeOrder=Ascending&delimiter=Semicolon&CIVILSTAND=U%2CG&ALDER=*&Tid=2020K4"
    df = get_data_frame(url)

    max_age = 125
    no_married_per_age = {}
    no_never_married_per_age = {}
    
    for index in range(1, max_age + 2):
        age = index - 1
        never_married = df.iloc[index, 3]
        married = df.iloc[index + max_age + 2, 3]
        
        no_married_per_age[age] = married
        no_never_married_per_age[age] = never_married
    
    ages = list(no_married_per_age.keys())
    married_per_age = list(no_married_per_age.values())
    never_married_per_age = list(no_never_married_per_age.values())
    
    plt.xlabel("Alder", fontsize = 10)
    plt.ylabel("Antal", fontsize = 10)
    plt.tick_params(axis = 'both', which = 'major', labelsize = 10)
    
    plt_married = plt.bar(ages, married_per_age, width = 0.5, linewidth = 0, align = "center", color = "red")
    plt_never_married = plt.bar(ages, never_married_per_age, width = 0.5, linewidth = 0, align = "center", color = "blue")
    
    plt.legend([plt_married, plt_never_married], ["Gift", "Aldrig gift"], loc = 1)
    
    
# task_5_a()
# task_5_b()
# task_5_c()
# task_5_d()
```
